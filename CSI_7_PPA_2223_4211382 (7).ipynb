##DATA PREPROCESSING

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the excel dataset as a data frame
dataset=pd.read_excel('dataset.xlsx')
```

```python
dataset.info()
```

```python
# gain valuable insight about our dataframe
dataset.describe()
```

```python
# view the first five rows of the dataset to see what kind of data we're dealing with
dataset.head()
```

```python
# drop the first column 195 which contained all zero values
dataset=dataset.drop(columns=[195], axis=1)
dataset.head()
```

```python
# find the shape of the dataset
dataset.shape
```

```python
dataset=dataset.drop(columns=['ID'], axis=1)
dataset.head()
```

```python
# check if there is missing data
dataset.isnull().sum()
```

```python
#drop the rows with missing data for Meth_flow
cleaned_data=dataset.dropna()
```

```python
#confirm to see if the null values have been removed
cleaned_data.isnull().sum()
```

```python
# split the data into features and the target
features = cleaned_data.drop(columns=['H_ppm'])
target = cleaned_data['H_ppm']

#encountered an error when using minMaxScalar, had to convert the features to string
features.columns=features.columns.astype(str)
```

```python
# check the shape of the splitted data
features.shape
```

```python
# check the shape of the splitted data
target.shape
```

Data Normalization


```python
# import the libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

Scale the data

```python
#scale our dataset
scaler=MinMaxScaler()
cleaned_data.columns=cleaned_data.columns.astype(str)
scaled_data=pd.DataFrame(scaler.fit_transform(cleaned_data), columns=cleaned_data.columns)
scaled_data
```

```python
#split the data into features and target
X = scaled_data.drop(columns=['H_ppm'])
Y = scaled_data['H_ppm']
```

```python
# split the data into training data and test data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape
y_train.shape
```

## REGRESSION ANALYSIS AND ASSOCIATED METRICS

```python
#import the libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import r2_score
```

```python
#initialize our model
model = Sequential()
```

```python
#define our model
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))  # Increased the first hidden layer
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
#compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
#fit the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=10, verbose=0)
```

Evaluating our Model

```python
# Evaluate the model
train_loss = model.evaluate(X_train, y_train, verbose=0)
test_loss = model.evaluate(X_test, y_test, verbose=0)
```

```python
#print the evaluation result
print(f"Train Loss: {train_loss}")
print(f"Test Loss: {test_loss}")
```

```python
#Use our Model to Prdeict
train_prediction=model.predict(X_train)
test_prediction=model.predict(X_test)
```

```python
#r2 score of our model
train_r2 = r2_score(y_train, train_prediction)
test_r2 = r2_score(y_test, test_prediction)
```

```python
#print the outputs
print(f"Train R^2: {train_r2:.3f}")
print(f"Test R^2: {test_r2:.3f}")
```

Visualize the actual vs predicted Values

```python
# Plot the training and validation loss curves
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plotting actual values against predicted values
plt.scatter(y_test, test_prediction, alpha=0.5)
plt.title('True vs Predicted Values for H_ppm')=
plt.xlabel('True H_ppm')
plt.ylabel('Predicted H_ppm')
plt.show()
```

##Classification Analysis

```python
#Import Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve,auc
from sklearn.preprocessing import label_binarize
```

```python
# Binarize the target variable for classification where 'H_ppm' > 0
y_class_train = (y_train > 0).astype(int)
y_class_test = (y_test > 0).astype(int)
```

```python
# Define the classification model
classifier = LogisticRegression()
```

```python
# Train the model
classifier.fit(X_train, y_class_train)
```

```python
# Predict on training and testing data
train_preds = classifier.predict(X_train)
test_preds = classifier.predict(X_test)
```

```python
# Evaluate the model
train_accuracy = accuracy_score(y_class_train, train_preds)
test_accuracy = accuracy_score(y_class_test, test_preds)
print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Detailed classification report
print(classification_report(y_class_test, test_preds))
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_class_test, test_preds)
print("Confusion Matrix:\n", conf_matrix)
```

```python
# ROC AUC Score
roc_auc = roc_auc_score(label_binarize(y_class_test, classes=[0, 1]), label_binarize(test_preds, classes=[0, 1]))
print(f"ROC AUC Score: {roc_auc}")
```

Visualize the Performance of the Model

```python
# Visualization 1: Confusion Matrix
conf_matrix = confusion_matrix(y_class_test, test_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Hydrogen', 'Hydrogen'], yticklabels=['No Hydrogen', 'Hydrogen'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Visualization 2: Class Distribution in Predicted Results
plt.figure(figsize=(8, 6))
sns.countplot(x=test_preds)
plt.title('Distribution of Predicted Classes')
plt.xticks(ticks=np.arange(2), labels=['No Hydrogen', 'Hydrogen'])
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```
